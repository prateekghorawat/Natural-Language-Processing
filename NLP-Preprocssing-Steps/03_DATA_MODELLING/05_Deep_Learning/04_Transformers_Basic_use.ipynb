```python
from transformers import pipeline
```

### **Sentiment Analysis**

```python
sa_model = pipeline("sentiment-analysis")
text = "Hi how are you doing today , is there anything i can hep with ?"

sa_model(text)
```

### **Classification**

```python
classification = pipeline("zero-shot-classification")
text = "I am learning NLP and trying to build some project as well"

classification(text ,
               candidate_labels = ["education" , "work" , "topic"])
```

### **Text Generation**

```python
generator = pipeline("text-generation")
generator("this section is all about transformer and")
```

### **Unmasking**

```python
unmask = pipeline("fill-mask")
unmask("This section will teach you all about <mask> models.", top_k=2)
```

### **Question Answering**

```python
question_answerer = pipeline("question-answering")
question_answerer(
    question="Where do I study?",
    context="My name is Prateek and I am pursing my studies from Schmakaden",
)
```

### **Named Entity Recognition**

```python
ner = pipeline("ner", grouped_entities=True)
ner("My name is Prateek and I am searching for a job, can you refer me.")
```
