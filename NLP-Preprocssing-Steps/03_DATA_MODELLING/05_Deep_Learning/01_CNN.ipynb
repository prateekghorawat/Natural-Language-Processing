```python
import tensorflow as tf
from tensorflow.keras.layers import (Conv1D , MaxPool1D , Flatten , Dense , Embedding)
from tensorflow.keras.models import Sequential
```

```python
## Maximum number of words or Size of Vocabulary
Max_Words_In_Vocab = 20000

## Dimension of Dense Embedding
Emb_Dim = 100

## Length of Input Sequence
Max_Seq_Len = 300

model = Sequential([
    Embedding(Max_Words_In_Vocab , Emb_Dim , input_length = Max_Seq_Len),
    Conv1D(128 , 5 , activation = "relu" ),
    MaxPool1D(5) ,
    Flatten(),
    Dense(2 , activation = "softmax")]
)
```

```python
model.compile (
loss = "categorical_crossentropy" ,
optimizer = "rmsprop" ,
metrics =["accuracy" ]
)
print ( model.summary () )
```
