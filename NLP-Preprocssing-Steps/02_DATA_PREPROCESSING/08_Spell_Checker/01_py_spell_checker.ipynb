```python
from spellchecker import SpellChecker
```

```python
spell = SpellChecker()
```

```python
mispelled = spell.unknown([
    "misspell" , "crazzy" , "prablem"
])
```

```python
for word in mispelled :
  print(f"{word} --> {spell.correction(word)}")
```
