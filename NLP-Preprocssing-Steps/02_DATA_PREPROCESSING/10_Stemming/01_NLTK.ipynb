```python
import nltk
from nltk.stem import PorterStemmer
from nltk import word_tokenize
nltk.download('punkt')
```

```python
text = "car cards cars automobile car computer compose fabrication"
token = word_tokenize(text)
```

```python
st = PorterStemmer ()
stemmed = " " . join ([ st.stem(word) for word in token ])
print(stemmed)
```
