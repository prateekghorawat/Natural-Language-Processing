```python
import sklearn
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
list1 = ["A" , "B" , "C" , "A" , "C" , "B" , "F" , "D" , "F"]
list1_ec = le.fit_transform(list1)
```

```python
print(f"{list1} \n{list1_ec}")
```

```python
for i in range(len(list1)) :
  print(f"{list1[i]} : {list1_ec[i]}")
```
