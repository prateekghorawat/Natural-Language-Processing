```python
import cv2
import numpy as np
from pytesseract import image_to_string
```

```python
filename = "/content/drive/MyDrive/Colab Notebooks/NLP_Specialization/Data/image.jpg"
img = cv2.imread(filename)
```

```python
def get_grayscale(image):
  return cv2.cvtColor(image , cv2.COLOR_BGR2GRAY)

def thresholding ( image ) :
  return cv2 . threshold (image , 0 , 255 , cv2 . THRESH_BINARY + cv2 . THRESH_OTSU) [1]

def opening ( image ) :
  return cv2 . morphologyEx (image , cv2 . MORPH_OPEN , np . ones ((5 ,5) , np . uint8 ))
```

```python
gray = get_grayscale(img)
thresh = thresholding(gray)
open = opening(thresh)

text_file = image_to_string(thresh , lang = "eng")
print(text_file)
```
