```python
import PyPDF2
from PyPDF2 import PdfReader
```

```python
print(PyPDF2.__version__)
```

```python
with open("/content/drive/MyDrive/Colab Notebooks/NLP_Specialization/Data/sample_merged.pdf" , "rb" ) as pdf:
  pdf_file_reader = PdfReader(pdf)
  num_pages = len(pdf_file_reader.pages)
  print(f"Total Number of Pages are : {num_pages}")
  i = 0
  if num_pages > 0 :
    while i < num_pages :
      page = pdf_file_reader.pages[2]
      text = page .extract_text()
      print(f"\nContent of the page : {i+1} \n")
      print(text)
      i = i + 1
```
